```python
import requests
import csv
import time
from bs4 import BeautifulSoup

# Danh sÃ¡ch cÃ´ng ty vÃ  mÃ£ tÆ°Æ¡ng á»©ng
companies = {
    "HoÃ ng Anh Gia Lai": "hag",
    "Vinamilk": "vnm",
    "HÃ²a PhÃ¡t": "hpg",
    "FPT": "fpt",
    "MB Bank": "mbb"
}

years = list(range(2019, 2025))

# HÃ m táº£i ná»™i dung vá»›i retry
def get_with_retry(url, max_retries=5, wait=3):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0 Safari/537.36"
    }

    for attempt in range(1, max_retries + 1):
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            print(f"    âœ… Táº£i thÃ nh cÃ´ng sau {attempt} láº§n thá»­")
            return response
        except requests.exceptions.HTTPError as e:
            print(f"    âš ï¸ Lá»—i HTTP {response.status_code} - thá»­ láº¡i sau {wait} giÃ¢y (láº§n {attempt})")
        except requests.exceptions.RequestException as e:
            print(f"    âŒ Lá»—i káº¿t ná»‘i: {e} - thá»­ láº¡i sau {wait} giÃ¢y (láº§n {attempt})")

        time.sleep(wait)

    print("    âŒ KhÃ´ng thá»ƒ táº£i trang sau nhiá»u láº§n thá»­.")
    return None

# Báº¯t Ä‘áº§u xá»­ lÃ½ tá»«ng cÃ´ng ty
for company_name, ticker in companies.items():
    print(f"ðŸ” Äang xá»­ lÃ½ cÃ´ng ty: {company_name} ({ticker})")

    final_table = []

    for year_index, year in enumerate(years):
        url = f'https://cafef.vn/du-lieu/bao-cao-tai-chinh/{ticker}/bsheet/{year}/4/0/0/bao-cao-tai-chinh-.chn'
        print(f"  -> NÄƒm {year}")

        response = get_with_retry(url)
        if response is None:
            continue  # bá» qua nÄƒm nÃ y náº¿u khÃ´ng táº£i Ä‘Æ°á»£c

        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table', id='tableContent')

        if not table:
            print(f"    âŒ KhÃ´ng tÃ¬m tháº¥y báº£ng dá»¯ liá»‡u")
            continue

        rows = table.find_all('tr')
        current_year_data = []

        for i, row in enumerate(rows):
            cols = row.find_all('td')
            if not cols:
                continue

            if len(cols) > 8:
                cols = cols[:-8]
            else:
                continue

            row_data = [col.text.strip() for col in cols]

            if len(row_data) < 2:
                continue

            if i == 0:
                if year_index == 0:
                    for q in range(1, len(row_data)):
                        if row_data[q] == '':
                            row_data[q] = f"QuÃ½ {q} nÄƒm {year}"
                    final_table.append(row_data)
                else:
                    new_titles = [f"QuÃ½ {q} nÄƒm {year}" for q in range(1, len(row_data))]
                    final_table[0].extend(new_titles)
            else:
                if year_index == 0:
                    current_year_data.append(row_data)
                else:
                    current_year_data.append(row_data[1:])  # bá» cá»™t Ä‘áº§u (chá»‰ tiÃªu)

        if year_index == 0:
            final_table.extend(current_year_data)
        else:
            for i in range(1, len(final_table)):
                final_table[i].extend(current_year_data[i - 1])

    # Ghi ra file CSV
    output_file = f"Tá»•ng há»£p bÃ¡o cÃ¡o {company_name}.csv"
    with open(output_file, mode='w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerows(final_table)

    print(f"âœ… ÄÃ£ lÆ°u vÃ o file: {output_file}\n")
```

```python
import requests
import csv
import time
from bs4 import BeautifulSoup

# Danh sÃ¡ch cÃ´ng ty vÃ  mÃ£ tÆ°Æ¡ng á»©ng
companies = {
    # "HoÃ ng Anh Gia Lai": "hag",
    # "Vinamilk": "vnm",
    # "HÃ²a PhÃ¡t": "hpg",
     "FPT": "fpt",
    # "MB Bank": "mbb"
}

years = list(range(2019, 2025))

# HÃ m táº£i ná»™i dung vá»›i retry
def get_with_retry(url, max_retries=5, wait=3):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0 Safari/537.36"
    }

    for attempt in range(1, max_retries + 1):
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            print(f"    âœ… Táº£i thÃ nh cÃ´ng sau {attempt} láº§n thá»­")
            return response
        except requests.exceptions.HTTPError as e:
            print(f"    âš ï¸ Lá»—i HTTP {response.status_code} - thá»­ láº¡i sau {wait} giÃ¢y (láº§n {attempt})")
        except requests.exceptions.RequestException as e:
            print(f"    âŒ Lá»—i káº¿t ná»‘i: {e} - thá»­ láº¡i sau {wait} giÃ¢y (láº§n {attempt})")

        time.sleep(wait)

    print("    âŒ KhÃ´ng thá»ƒ táº£i trang sau nhiá»u láº§n thá»­.")
    return None

# Báº¯t Ä‘áº§u xá»­ lÃ½ tá»«ng cÃ´ng ty
for company_name, ticker in companies.items():
    print(f"ðŸ” Äang xá»­ lÃ½ cÃ´ng ty: {company_name} ({ticker})")

    final_table = []

    for year_index, year in enumerate(years):
        url = f'https://cafef.vn/du-lieu/bao-cao-tai-chinh/{ticker}/incsta/{year}/4/0/0/bao-cao-tai-chinh-.chn'
        print(f"  -> NÄƒm {year}")

        response = get_with_retry(url)
        if response is None:
            continue  # bá» qua nÄƒm nÃ y náº¿u khÃ´ng táº£i Ä‘Æ°á»£c

        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table', id='tableContent')

        if not table:
            print(f"    âŒ KhÃ´ng tÃ¬m tháº¥y báº£ng dá»¯ liá»‡u")
            continue

        rows = table.find_all('tr')
        current_year_data = []

        for i, row in enumerate(rows):
            cols = row.find_all('td')
            if not cols:
                continue

            # Bá» cá»™t cuá»‘i vÃ  cá»™t Ä‘áº§u tiÃªn nhÆ° trong code cÅ©
            if len(cols) > 8:
                cols = cols[:-8]
            else:
                continue

            row_data = [col.text.strip() for col in cols]

            if len(row_data) < 2:
                continue

            if i == 0:
                if year_index == 0:
                    # Láº§n Ä‘áº§u tiÃªn, thÃªm quÃ½ cho dÃ²ng tiÃªu Ä‘á» Ä‘áº§u tiÃªn
                    for q in range(1, len(row_data)):
                        if row_data[q] == '':
                            row_data[q] = f"QuÃ½ {q} nÄƒm {year}"
                    final_table.append(row_data)  # ThÃªm vÃ o báº£ng cho nÄƒm Ä‘áº§u
                else:
                    # KhÃ´ng thay Ä‘á»•i dÃ²ng tiÃªu Ä‘á» cho cÃ¡c nÄƒm tiáº¿p theo
                    pass
            else:
                if year_index == 0:
                    current_year_data.append(row_data)
                else:
                    # Bá» cá»™t Ä‘áº§u tiÃªn khi ghÃ©p nÄƒm tiáº¿p theo
                    current_year_data.append(row_data[1:])

        # Chá»‰ thÃªm dá»¯ liá»‡u nÄƒm sau vÃ o final_table, giá»¯ nguyÃªn dá»¯ liá»‡u cá»§a nÄƒm Ä‘áº§u
        if year_index == 0:
            final_table.extend(current_year_data)
        else:
            for i in range(1, len(final_table)):
                final_table[i].extend(current_year_data[i - 1])

    # Táº¡o new_row chá»©a thÃ´ng tin tá»« QuÃ½ 1 nÄƒm 2019 Ä‘áº¿n QuÃ½ 4 nÄƒm 2024
    new_row = ['']  # Äá»ƒ cá»™t Ä‘áº§u tiÃªn trá»‘ng
    for year in years:
        for q in range(1, 5):
            new_row.append(f"QuÃ½ {q} nÄƒm {year}")

    # ThÃªm new_row vÃ o trÆ°á»›c dÃ²ng Ä‘áº§u tiÃªn (chá»‰ thÃªm má»™t láº§n)
    final_table.insert(0, new_row)

    # Ghi ra file CSV
    output_file = f"Tá»•ng há»£p bÃ¡o cÃ¡o {company_name} 2.csv"
    with open(output_file, mode='w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerows(final_table)

    print(f"âœ… ÄÃ£ lÆ°u vÃ o file: {output_file}\n")
```
