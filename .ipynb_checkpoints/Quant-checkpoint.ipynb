```python
import pandas as pd

# Đọc dữ liệu từ CSV (bạn có thể đổi thành Excel hoặc DataFrame khác)
df = pd.read_csv("MB Bank_stock_data.csv")

# Bỏ 3 cột cuối
df = df.iloc[:, :6]

# Đổi tên cột nếu cần (nếu bạn chưa đặt tên)
df.columns = ['date', 'col2', 'col3', 'col4', 'col5', 'col6']

# Chuyển cột date sang định dạng datetime
df['date'] = pd.to_datetime(df['date'], utc=True)

# Gán quý cho mỗi dòng
df['quarter'] = df['date'].dt.to_period('Q')

# Nhóm theo quý
grouped = df.groupby('quarter').agg({
    'col2': 'mean',
    'col3': 'mean',
    'col4': 'mean',
    'col5': 'mean',
    'col6': 'sum'
}).reset_index()

# Đổi tên cột nếu cần
grouped.columns = ['quarter', 'avg_col2', 'avg_col3', 'avg_col4', 'avg_col5', 'sum_col6']

# In kết quả
print(grouped)

```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd

# Giả sử `grouped` đã có, và bạn đã bỏ cột 'quarter'
data = grouped.drop(columns=['quarter']).reset_index(drop=True)

rows = []
for i in range(len(data) - 2):
    row1 = data.iloc[i].add_suffix('_1')
    row2 = data.iloc[i + 1].add_suffix('_2')
    target = data.iloc[i + 2]['sum_col6']
    
    combined = pd.concat([row1, row2])
    combined['target'] = target
    combined['index'] = i + 1  # Số thứ tự
    
    rows.append(combined)

# Chuyển danh sách các series thành DataFrame
final_df = pd.DataFrame(rows)

# Đưa 'index' lên đầu nếu muốn
cols = ['index'] + [col for col in final_df.columns if col != 'index']
final_df = final_df[cols]

print(final_df)
```

```python
X = final_df.drop(columns=['index', 'target'], axis=1)
Y = final_df['target']
X, Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Khởi tạo và huấn luyện mô hình
model = LinearRegression()
model.fit(X_train, Y_train)

# Dự đoán
Y_pred = model.predict(X_test)

# Đánh giá mô hình
print("Mean Squared Error (MSE):", mean_squared_error(Y_test, Y_pred))
print("R^2 Score:", r2_score(Y_test, Y_pred))

# (Tùy chọn) Hiển thị một vài kết quả dự đoán
results = X_test.copy()
results['Actual'] = Y_test.values
results['Predicted'] = Y_pred
print(results.head())
```

```python
import matplotlib.pyplot as plt

# Nếu có cột 'quarter' là thời gian, hãy sắp xếp và vẽ
plot_df = final_df.copy()
plot_df['quarter'] = grouped['quarter'].iloc[2:].reset_index(drop=True)  # lấy cột thời gian phù hợp

# Đảm bảo cột thời gian ở dạng string hoặc datetime để plot
plot_df['quarter'] = plot_df['quarter'].astype(str)

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(plot_df['quarter'], plot_df['target'], marker='o', linestyle='-')
plt.title('Target value over time')
plt.xlabel('Quarter')
plt.ylabel('Target (col6 của hàng kế tiếp)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Giả sử df_quarter là dataframe đã tính trung bình và tổng theo quý (grouped)
# Gồm các cột: quarter, col2, col3, col4, col5, col6

fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle("So sánh các cột trung bình với col6 theo quý", fontsize=16)

cols = ['avg_col2', 'avg_col3', 'avg_col4', 'avg_col5']
quarters = grouped['quarter'].astype(str)
col6 = grouped['sum_col6']

for i, col in enumerate(cols):
    ax = axs[i // 2, i % 2]
    ax.plot(quarters, grouped[col], label=col, marker='o')
    ax.plot(quarters, col6, label='col6 (sum)', marker='x')
    ax.set_title(f'{col} vs col6')
    ax.set_xlabel('Quarter')
    ax.set_ylabel('Value')
    ax.legend()
    ax.grid(True)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
