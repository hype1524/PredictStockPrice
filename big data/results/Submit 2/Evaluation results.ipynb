```python
import pandas as pd
import os
from glob import glob

# Bước 1: Đọc tất cả các file CSV
file_paths = glob("results_*.csv")

# Danh sách lưu kết quả trung bình
results = []

for file_path in file_paths:
    df = pd.read_csv(file_path)
    model_name = os.path.splitext(os.path.basename(file_path))[0].replace("results_", "")
    df['model'] = model_name

    # Lọc loss_function == "mse" và chỉ lấy activation là relu hoặc gelu
    df = df[(df['loss_function'] == 'mse') & (df['activation'].isin(['relu', 'gelu', 'linear']))]

    # Tính trung bình final_loss theo activation
    mean_losses = df.groupby('activation')['final_loss'].mean().to_dict()

    # Gộp kết quả vào 1 dict theo model
    results.append({
        'model': model_name,
        'relu': mean_losses.get('relu', None),
        'gelu': mean_losses.get('gelu', None),
        'linear': mean_losses.get('linear', None)
    })

# Tạo DataFrame kết quả cuối cùng
summary_df = pd.DataFrame(results)


summary_df
```
