```python
import pandas as pd

# Đọc file CSV
df = pd.read_csv("results_triple_stacked_lstm.csv")

# Lọc theo activation = relu và loss_function = mse
filtered_df = df[(df['activation'] == 'relu') & (df['loss_function'] == 'mse')]

# Nhóm theo n_steps_in và n_steps_out rồi tính trung bình final_loss
loss_by_steps = filtered_df.groupby(['a', 'b'])['final_loss'].mean().reset_index()

# Đổi tên cột cho dễ hiểu khi dùng vẽ biểu đồ hoặc đưa vào bảng
loss_by_steps.rename(columns={
    'a': 'n_steps_in',
    'b': 'n_steps_out',
    'final_loss': 'final_loss_relu_mse'
}, inplace=True)

# Hiển thị kết quả
loss_by_steps
```
